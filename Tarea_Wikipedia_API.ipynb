```python
import pandas as pd
import requests
```

```python
#- Extraer los resúmenes de artículos de wikipedia que contengan la palabra 'astro' (limitar a máximo 20)
palabra = 'astro'
endpoint = 'https://es.wikipedia.org/w/api.php'

parametros = {
    'action':'query',
    'format':'json',
    'list':'search',
    'srsearch': palabra,
    'prop':'extracts',
    'srlimit':20
            }

wiki_req = requests.get(endpoint, params=parametros)
print(wiki_req.status_code)

df = pd.json_normalize(wiki_req.json()['query']['search'])
df
```

```python
#DataFrame con columnas :pageid,titulo, resumen, contien_palabra
import pandas as pd

palabra ='astro'
endpoint ='https://es.wikipedia.org/w/api.php'

params = {
    'action': 'query',
    'format': 'json',
    'list': 'search',
    'srsearch': palabra,
    'prop': 'extracts',
    'srlimit': 20
}

```

```python
#- Crear una variable con 0 o 1 que indique si el resumen contiene alguna de las palabras 'física' o 'ciencia'
wiki_req = requests.get(endpoint, params=params)
print(wiki_req.status_code)

resultados = wiki_req.json()['query']['search']

resumenes = []
contar_palabra = []

for result in resultados:
    params = {
        'action': 'query',
        'format': 'json',
        'prop': 'extracts',
        'exintro': True,
        'explaintext': True,
        'pageids': result['pageid']
    }
    wiki_req = requests.get(endpoint, params=params)
    result = wiki_req.json()['query'].get('pages', {})
    resumen = list(result.values())[0].get('extract', '')

    contar_fisica = int('fisica' in resumen.lower())
    contar_ciencia = int('ciencia' in resumen.lower())

    resumenes.append(resumen)
    contar_palabra.append(contar_fisica or contar_ciencia)

df = pd.DataFrame({
    'pageid': [result['pageid'] for result in resultados],
    'titulo': [result['title'] for result in resultados],
    'resumen': resumenes,
    'contiene_palabra': contar_palabra
})

df
```
