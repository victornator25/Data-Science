```python
from pickle import FALSE
import matplotlib.pyplot as plt
from PIL import Image

#cambiar de fondo a una imagen
imagen = Image.open('lunallena.jpg')
imagen.mode
data=imagen.load()
largo, ancho = imagen.size

for i in range(largo):
  for j in range(ancho):
    (r,g,b)= data[i,j]
    if (r,g,g) == (0,0,0):
      data[i,j]= (255, 0, 255)

fig,axs = plt.subplots(figsize=(6,6))
axs.imshow(imagen)
plt.show()
```

```python
import numpy as np
imagen2 = Image.open('cheems.jpg')
im = np.array(imagen2)
imagen2
```

```python
for i in range(im.shape[0]):
    for j in range(im.shape[1]):
      pixel = im[i][j]
      promedio = (pixel[0] + pixel[1] + pixel[2])//3
      im[i][j][0] = promedio
      im[i][j][1] = promedio
      im[i][j][2] = promedio

fig_3,axs_3 = plt.subplots(figsize=(6,6))
axs_3.imshow(im)
plt.show()
```
